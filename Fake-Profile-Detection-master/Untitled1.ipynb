```python
dataset=pd.read_csv('Total_data.csv')
```

```python
y=dataset.FakeOrNot
```

```python
x=dataset.drop(columns=['FakeOrNot'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_tesfromt=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_log=LogisticRegression().fit(x_train,y_train)
```

```python
pred=model_log.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_fire=DecisionTreeClassifier().fit(x_train,y_train)
```

```python
pred=model_fire.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
from sklearn.svm import SVC
```

```python
model_ashik=SVC().fit(x_train,y_train)
```

```python
pred=model_ashik.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```
